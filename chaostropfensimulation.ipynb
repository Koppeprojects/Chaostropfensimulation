Der nächste Block führt eine Funktion ein die basierend auf der momentanen $position$ etc. die $position$ etc. im nächsten Zeitschritt berechnet. Das ist wie wenn man ein Video Bild für Bild anschaut und jeweils das neue aus dem alten vorausberechnet. 

```python
def Zeitschritt_gehen(zeitschrittdauer,position,geschwindigkeit,beschleunigung):
    position_neu=position+geschwindigkeit*zeitschrittdauer+ beschleunigung/2*zeitschrittdauer**2
    geschwindigkeit_neu=geschwindigkeit+beschleunigung*zeitschrittdauer
    return position_neu, geschwindigkeit_neu
```

Die Funktion im nächsten Block speichert die momentane $position$ etc. in den Speicher damit wir später darauf zugreifen können.

```python
def Speichern(zeitschritt,position,geschwindigkeit,riss,position_speicher,geschwindigkeit_speicher,riss_speicher):
    position_speicher[zeitschritt]=position
    geschwindigkeit_speicher[zeitschritt]=geschwindigkeit
    riss_speicher[zeitschritt]=riss
    return position_speicher,geschwindigkeit_speicher,riss_speicher
```

```python
#Graphen von den gespeicherten Daten der Simulation
fig, ax = plt.subplots(figsize=(14, 7))
ax.plot(zeit_speicher,position_speicher)
ax.set_xlabel('zeit in s')
ax.set_ylabel('position in mm')
plt.show()
```
